```python
zip_name ="readonly/small_img.zip"

# list holds the images names
imagename_list=[]
# dictionlay holds the image and pixel info
images_dict = {} # key image name and value list (image and its details)

def extract_images(zip_name):
    # opening the zip file in READ mode 
    with zp.ZipFile(zip_name, 'r') as zip: 
        
        for f in zip.infolist():
            
            images_dict[f.filename] = [Image.open(zip.open(f.filename))]
            imagename_list.append(f.filename)
    
# call method to extract images from zip
extract_images(zip_name)

for imgname in imagename_list:
    
    #print(imgname)
    #display(images_dict[imgname][0])
    
    img = images_dict[imgname][0]
    #pytesseract.image_to_string(img)
    
    # append the extracted text from image to the referrenced dict. list value 
    images_dict[imgname].append(pytesseract.image_to_string(img))
    
    # check 'Christopher' exists in the extracted text from the image
    #if 'Christopher' in images_dict[imgname][1]:
    #    print("Results found in file {}".format(imgname))
    
    # check 'Mark' exists in the extracted text from the image
    if 'Mark' in images_dict[imgname][1]:
        
        print("Results found in file {}".format(imgname))
    
    try:
        # check for the faces in the image
        faces = (face_cascade.detectMultiScale(np.array(img),1.35,4)).tolist()

        # append the detected faces to the refferenced dict. list value
        images_dict[imgname].append(faces)

        faces_in_each = []

        for x,y,w,h in images_dict[imgname][2]:
            faces_in_each.append(img.crop((x,y,x+w,y+h)))

        contact_sheet = Image.new(img.mode, (550,110*int(np.ceil(len(faces_in_each)/5))))

        x = 0
        y = 0

        for face in faces_in_each:

            face.thumbnail((110,110))

            contact_sheet.paste(face, (x, y))

            if x+110 == contact_sheet.width:
                x=0
                y=y+110
            else:
                x=x+110
                
        display(contact_sheet)
        
    except:

        print('But there were no faces in that file!')
    
     
```
